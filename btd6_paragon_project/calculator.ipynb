```python
# @title Paragon Inputs
difficulty = 'Medium' # @param ['Easy','Medium','Hard','Impoppable']
paragon_name = "Glaive Dominus" # @param ["Apex Plasma Master","Glaive Dominus","Ascended Shadow","Navarch of the Seas","Goliath Doomship"]
# @markdown number of additional tier 5 towers, not including the first 3 required to make the paragon
tier_5_towers = 0 # @param {"type":"slider","min":0,"max":9,"step":1}
# @markdown upgrades refers to the number of upgrades on non-tier 5 towers
upgrades = 0 # @param {"type":"slider","min":0,"max":100,"step":1}
# @markdown Cash injection is the bonus cash spent when upgrading the paragon.
cash_injection = 0 # @param {"type":"number"}
# @markdown Exclude the 3 cheapest tier 5 towers
cost_of_sacrificed_towers = 0 # @param {"type":"number"}
# @markdown Total pops and income for all sacrificed towers
pops = 0 # @param {"type":"slider","min":0,"max":16200000,"step":10000}
income = 0 # @param {"type":"slider","min":0,"max":4050000,"step":10000}
# @markdown If using Geraldo, input the number of paragon power totems deployed
Geraldo_power_totems = 0 # @param {"type":"slider","min":0,"max":100,"step":1}
legacy_output = False # @param {"type":"boolean"}

## form ends, code starts
os.chdir("/content/drive/MyDrive/btd6_paragon_project")
with open('p2d.csv','r') as f:
  p2ddf = pd.read_csv(f)
p2ddf.set_index('degree')

with open('Paragon Prices.csv','r') as f:
  pricesdf = pd.read_csv(f)
pricesdf.set_index('Difficulty',inplace=True)
paragon_price = pricesdf.loc[difficulty,paragon_name]
pharold = paragon_price/20000
geraldoTotemCostList = {'Easy':22100,'Medium':26000,'Hard':28080,'Impoppable':31200}
geraldo_totem_cost = geraldoTotemCostList.get(difficulty)

# calculate power
tier5power = min(50000,tier_5_towers*6000)
upgradeNumPower = min(10000,100*upgrades)
upgradeCostPower = min(60000,(cost_of_sacrificed_towers/pharold)+(cash_injection/(1.05*pharold)))
popsIncomePower = min(90000,(pops/180)+(income/45))
totemPower = min(200000,(Geraldo_power_totems*2000))

power = min(200000,round(tier5power+upgradeNumPower+upgradeCostPower+popsIncomePower+totemPower))

max_injection = round(paragon_price*3.15)
max_impactful_injection = round((1.05*pharold)*(60000-(cost_of_sacrificed_towers/pharold)))

## final degree calculation

if power>=200000:
  degree = 100
else:
  for x in range(1,100):
    degree = x
    checkpower = p2ddf['power'][x]
    if checkpower>power:
      break

full_cost = paragon_price+cash_injection+(geraldo_totem_cost*Geraldo_power_totems)+cost_of_sacrificed_towers

display(Markdown(f"#Power Stats:"))
display(Markdown(f"Power from tier 5 towers: **{round(tier5power)}**/50000"))
display(Markdown(f"Power from non-tier 5 upgrades: **{round(upgradeNumPower)}**/10000"))
display(Markdown(f"Power from upgrade costs & cash injection: **{round(upgradeCostPower)}**/60000"))
display(Markdown(f"Power from pops & income: **{round(popsIncomePower)}**/90000"))
display(Markdown(f"Power from power totems: **{totemPower}**"))
print('')
display(Markdown(f"Total power: {power}"))
display(Markdown(f"Degree: **{degree}**"))
if degree<100:
  display(Markdown(f"Power for next degree: {checkpower}"))
print('')
display(Markdown(f"Base Price: {paragon_price}"))
display(Markdown(f"Cost with current injection: **{paragon_price+cash_injection}**"))
display(Markdown(f"Total cost including totems and sacrificed towers: **{full_cost}**"))
print('')
## analysis (ie stats per cost)
display(Markdown(f"Max cash injection: {max_injection}"))
display(Markdown(f"Max impactful cash injection: **{max_impactful_injection}**"))

print('')

# injX, injY = injection_graph(max_impactful_injection,tier5power+upgradeNumPower+popsIncomePower+totemPower,)

# plt.plot(injX,injY,color='red')
# plt.show()

## get stats for paragon at calculated degree
os.chdir("/content/drive/MyDrive/btd6_paragon_project/Paragon_Stats")
csvName = paragon_name+'.csv'
with open(csvName,'r') as f:
  statsdf = pd.read_csv(f)
statsdf.set_index('degree',inplace=True)

degreestats = statsdf.iloc[degree-1]

# print(degreestats)

# display(Markdown(f""))

# basically the idea is this...

display(Markdown(f"power per $1: {round(power/full_cost,4)}"))
if degree/full_cost==0:
  display(Markdown(f"degree per $1,000: **0**"))
elif degree/full_cost<0.0001:
  display(Markdown(f"degree per $1,000: **<0.0001**"))
else:
  display(Markdown(f"degree per $1,000: **{round(degree/full_cost,4)}**"))
print('')
display(Markdown(f"#Paragon Stats:"))

if legacy_output:
  display(Markdown(f"##legacy output"))
  for x in statsdf.keys():
    display(Markdown(f"{x}:  **{statsdf[x][degree]}**"))
    # display(Markdown(f"{x} per $1000: {(statsdf[x][degree]*1000)/full_cost}"))

  # maindps = round((statsdf['main attack damage'][degree]*statsdf['main attack projectile count'][degree])/(statsdf['main attack cooldown'][degree]),3)
  # display(Markdown(f"main attack dps: **{maindps}**"))
  # display(Markdown(f"main attack dps per $1,000: **{(maindps*1000)/(full_cost)}**"))
else:
  if paragon_name=='Glaive Dominus':
    glaive_dominus_display(degreestats)
  elif paragon_name=='Navarch of the Seas':
    navarch_of_the_seas_display(degreestats)
  elif paragon_name=='Goliath Doomship':
    goliath_doomship_display(degreestats)
  else:
    display(Markdown(f"##legacy output"))
    for x in statsdf.keys():
      display(Markdown(f"{x}:  **{statsdf[x][degree]}**"))
```
